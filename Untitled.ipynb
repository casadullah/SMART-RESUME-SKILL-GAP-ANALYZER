```python
# ==========================================
# SMART RESUME SKILL GAP ANALYZER
# ==========================================

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Sample resume and job description
resume_text = """
Python SQL Excel Pandas Data Analysis Machine Learning
"""

job_description = """
Python SQL Power BI Machine Learning Deep Learning Pandas
"""

# Step 3: Define important skills list
skills_database = [
    "Python",
    "SQL",
    "Excel",
    "Pandas",
    "NumPy",
    "Machine Learning",
    "Deep Learning",
    "Power BI",
    "Tableau",
    "Statistics",
    "Data Visualization"
]

# Step 4: Extract skills
resume_skills = []
job_skills = []

for skill in skills_database:
    if skill.lower() in resume_text.lower():
        resume_skills.append(skill)

    if skill.lower() in job_description.lower():
        job_skills.append(skill)

# Step 5: Find matching and missing skills
matched_skills = list(set(resume_skills) & set(job_skills))
missing_skills = list(set(job_skills) - set(resume_skills))

# Step 6: Calculate match score
match_score = (len(matched_skills) / len(job_skills)) * 100

# Step 7: Print results
print("Resume Skills:", resume_skills)
print("Job Required Skills:", job_skills)
print("Matched Skills:", matched_skills)
print("Missing Skills:", missing_skills)
print(f"Match Percentage: {match_score:.2f}%")

# Step 8: Visualization
categories = ['Matched', 'Missing']
values = [len(matched_skills), len(missing_skills)]

plt.figure(figsize=(6,4))
plt.bar(categories, values)
plt.title("Skill Match Analysis")
plt.ylabel("Number of Skills")
plt.show()

# Step 9: Suggest improvements
print("\nRecommended skills to learn:")
for skill in missing_skills:
    print("-", skill)
```
